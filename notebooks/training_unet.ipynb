```python
import tensorflow as tf
import cv2
import numpy as np
import os
from models.unet_model import unet_model

IMG_DIR = "data/LEVIR-CD/patches/images"
MASK_DIR = "data/LEVIR-CD/patches/masks"

X, Y = [], []

for name in os.listdir(IMG_DIR):
    img = cv2.imread(os.path.join(IMG_DIR, name))
    mask = cv2.imread(os.path.join(MASK_DIR, name), 0)

    X.append(img / 255.0)
    Y.append(mask / 255.0)

X = np.array(X)
Y = np.expand_dims(np.array(Y), axis=-1)

model = unet_model(input_shape=(256, 256, 3))
model.fit(X, Y, epochs=10, batch_size=8, validation_split=0.1)

model.save("models/unet_model.h5")
```
